```python
import pandas as pd
import csv

import matplotlib.pyplot as plt
# Visualizations will be shown in the notebook.
%matplotlib inline
```

```python
with open('KP_tuning_value.txt', 'r') as in_file:
    stripped = (line.strip() for line in in_file)
    lines = (line.split(",") for line in stripped if line)
    with open('KP_tuning_value.csv', 'w') as out_file:
        writer = csv.writer(out_file)
        #writer.writerow(('kp', 'timestep', 'cte'))
        writer.writerows(lines)
        
columns=['Kp', 'time_step', 'cte'] 
df = pd.read_csv('KP_tuning_value.csv' , names = columns,header = None)       
```

```python
df1 = df.loc[df['Kp'] == -0.4]
df2 = df.loc[df['Kp'] == -0.3]
df3 = df.loc[df['Kp'] == -0.2]
df4 = df.loc[df['Kp'] == -0.1]
df5 = df.loc[df['Kp'] == 0.1]
df6 = df.loc[df['Kp'] == 0.2]
df7 = df.loc[df['Kp'] == 0.3]
df8 = df.loc[df['Kp'] == 0.4]
```

```python
kp_neg_0_4 =[]
kp_neg_0_3 =[]
kp_neg_0_2 =[]
kp_neg_0_1 =[]
kp_pos_0_1 =[]
kp_pos_0_2 =[]
kp_pos_0_3 =[]
kp_pos_0_4 =[]
time_step_label =[]

for i in range(1,401):
    time_step_label.append(i) 

for i, row in df1.iterrows():
    kp_neg_0_4.append(df.loc[i,'cte'])
    
for i, row in df2.iterrows():
    kp_neg_0_3.append(df.loc[i,'cte'])
    
for i, row in df3.iterrows():
    kp_neg_0_2.append(df.loc[i,'cte'])  
    
for i, row in df4.iterrows():
    kp_neg_0_1.append(df.loc[i,'cte']) 
       
for i, row in df5.iterrows():
    kp_pos_0_1.append(df.loc[i,'cte'])    

for i, row in df6.iterrows():
    kp_pos_0_2.append(df.loc[i,'cte']) 
    
for i, row in df7.iterrows():
    kp_pos_0_3.append(df.loc[i,'cte'])    

for i, row in df8.iterrows():
    kp_pos_0_4.append(df.loc[i,'cte'])    
```

```python
fig, ax = plt.subplots(figsize=(15,7))
ax.plot(time_step_label , kp_neg_0_4,label='-0.4',color = 'blue')
ax.plot(time_step_label , kp_neg_0_3,label='-0.3',color = 'blue')
ax.plot(time_step_label , kp_neg_0_2,label='-0.2',color = 'blue')
ax.plot(time_step_label , kp_neg_0_1,label='-0.1',color = 'blue')
ax.plot(time_step_label , kp_pos_0_1,label='0.1',color = 'red',lw=4)
ax.plot(time_step_label , kp_pos_0_2,label='0.2',color = 'black',lw=4)
ax.plot(time_step_label , kp_pos_0_3,label='0.3',color = 'green',lw=4)
ax.plot(time_step_label , kp_pos_0_4,label='0.4',color = 'blue')
ax.grid()
ax.set_title('Time_step vs Cte (Cross track error) - KP parameter Analysis', fontsize=15)
ax.set_xlabel('Time step',fontsize=15)
ax.set_ylabel('cte - Cross track error',fontsize=15)
ax.legend(loc='upper right')
ax.set_xlim((200, 400))
ax.set_ylim((-10, 10))


plt.savefig('KP_parameter_Decimal_Analysis.jpg')
plt.show()
```
